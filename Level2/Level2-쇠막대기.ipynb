```python
def solution(arrangement):
    cnt = 0
    arrangement = arrangement.replace('()', '|')
    stick = []
    
    for i in arrangement:
        
        if i == '(':
            stick.append(i)
        elif i == ')':
            stick.pop(-1)
            cnt += 1
        else:
            cnt += len(stick)
    
    return cnt
```

```python
arrangement = "()(((()())(())()))(())"
solution(arrangement)
```
